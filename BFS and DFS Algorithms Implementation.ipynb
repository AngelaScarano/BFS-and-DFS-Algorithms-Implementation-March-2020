```python
class Node:
    def __init__(self, parent, data):
        self.parent = parent
        self.data = data
        self.children = []
```

```python
#assume each jug, after size is defined, starts full
from queue import Queue
from queue import LifoQueue

def jugs (jugsize1, jugsize2, jugsize3, endstate1, endstate2, endstate3, algorithm):
    jug1 = jugsize1
    jug2 = jugsize2
    jug3 = jugsize3
    goal = (endstate1, endstate2, endstate3)
    
    if (algorithm == "bfs"):
        bfs(jug1, jug2, jug3, goal)
        
    if (algorithm == "dfs"):
        dfs(jug1, jug2, jug3, goal)
```

```python
#Uses a normal queue to implement bfs
def bfs (jug1, jug2, jug3, goal):
    #make a queue to store the in-progress bfs
    queue = Queue(0)
    
    #make a closed list to store where the bfs has been - use it to prevent repeated nodes
    #store tuples of node data values here
    closedList = []
    
    jugCap = [jug1, jug2, jug3]
    initJugStats = [jug1, jug2, jug3]
    initJugTuple = (jug1, jug2, jug3)
    
    #root to store starting point, and keep track of the bfs pathways
    treeRoot = Node(None, initJugTuple)
    
    queue.put(treeRoot)
    currJugStats = [jug1, jug2, jug3]
    currNode = treeRoot
    
    stepCounter = 0
    while queue.empty() == False:
        #update step counter
        stepCounter = stepCounter + 1
        
        #print('Here We Go!')
        currNode = queue.get()
        tupleCurr = (currNode.data[0], currNode.data[1], currNode.data[2])
        currJugStats = [currNode.data[0], currNode.data[1], currNode.data[2]]
        
        ##check if current node hits the goal
        if tupleCurr == goal:
            "Solution Found!"
            break
            
        ##check if current node is in closed list, stop that node from expanding if it is
        inList = False
        length = len(closedList)
        for i in range(length):
            if (tupleCurr == closedList[i]):
                continue
                        
        ##expand node with every jug change possible
        
        #empty 1st jug
        tempJugStats = []
        for i in range(0, 3):
            tempJugStats.append(currJugStats[i])
        tempJugStats[0] = 0
        tempTuple = (tempJugStats[0], tempJugStats[1], tempJugStats[2])
                
        #add the new jug state to the queue as a node
        newNode = Node(currNode, tempTuple)
        inList = False
        length = len(closedList)
        for i in range(length):
            if (tempTuple == closedList[i]):
                inList = True
        if inList == False:
            queue.put(newNode)
            currNode.children.append(newNode)
            
        tempJugStats.clear()
        
        #empty 2nd jug
        tempJugStats = []
        for i in range(0, 3):
            tempJugStats.append(currJugStats[i])
        tempJugStats[1] = 0
        tempTuple = (tempJugStats[0], tempJugStats[1], tempJugStats[2])
                
        #add the new jug state to the queue as a node
        newNode = Node(currNode, tempTuple)
        inList = False
        length = len(closedList)
        for i in range(length):
            if (tempTuple == closedList[i]):
                inList = True
        if inList == False:
            queue.put(newNode)
            currNode.children.append(newNode)
            
        tempJugStats.clear()
        
        #empty 3rd jug
        tempJugStats = []
        for i in range(0, 3):
            tempJugStats.append(currJugStats[i])
        tempJugStats[2] = 0
        tempTuple = (tempJugStats[0], tempJugStats[1], tempJugStats[2])
                
        #add the new jug state to the queue as a node
        newNode = Node(currNode, tempTuple)
        inList = False
        length = len(closedList)
        for i in range(length):
            if (tempTuple == closedList[i]):
                inList = True
        if inList == False:
            queue.put(newNode)
            currNode.children.append(newNode)
            
        tempJugStats.clear()
        
        #fill 1st jug
        tempJugStats = []
        for i in range(0, 3):
            tempJugStats.append(currJugStats[i])   
        tempJugStats[0] = jugCap[0]
        tempTuple = (tempJugStats[0], tempJugStats[1], tempJugStats[2])
        
        #add the new jug state to the queue as a node
        newNode = Node(currNode, tempTuple)
        inList = False
        length = len(closedList)
        for i in range(length):
            if (tempTuple == closedList[i]):
                inList = True
        if inList == False:
            queue.put(newNode)
            currNode.children.append(newNode)
            
        tempJugStats.clear()
        
        #fill 2nd jug
        tempJugStats = []
        for i in range(0, 3):
            tempJugStats.append(currJugStats[i])   
        tempJugStats[1] = jugCap[1]
        tempTuple = (tempJugStats[0], tempJugStats[1], tempJugStats[2])
        
        #add the new jug state to the queue as a node
        newNode = Node(currNode, tempTuple)
        inList = False
        length = len(closedList)
        for i in range(length):
            if (tempTuple == closedList[i]):
                inList = True
        if inList == False:
            queue.put(newNode)
            currNode.children.append(newNode)
            
        tempJugStats.clear()
        
        #fill 3rd jug
        tempJugStats = []
        for i in range(0, 3):
            tempJugStats.append(currJugStats[i])   
        tempJugStats[2] = jugCap[2]
        tempTuple = (tempJugStats[0], tempJugStats[1], tempJugStats[2])
        
        #add the new jug state to the queue as a node
        newNode = Node(currNode, tempTuple)
        inList = False
        length = len(closedList)
        for i in range(length):
            if (tempTuple == closedList[i]):
                inList = True
        if inList == False:
            queue.put(newNode)
            currNode.children.append(newNode)
            
        tempJugStats.clear()
        
        #check every water jug combination, see if water movement can be done
        tempJugStats = []
        for x in range(0, 3):
            for j in range(0, 3):
                for z in range(0, 3):
                    tempJugStats.append(currJugStats[z]) 
                #print('x: ', x)
                #print('j: ', j)
                
                if (x != j):
                    waterCanRecieve = jugCap[j] - tempJugStats[j]
                    waterCanGive = tempJugStats[x]
                    if (waterCanGive > waterCanRecieve):
                        tempJugStats[j] = tempJugStats[j] + waterCanRecieve
                        tempJugStats[x] = tempJugStats[x] - waterCanRecieve
                    elif(waterCanGive < waterCanRecieve):
                        tempJugStats[j] = tempJugStats[j] + waterCanGive
                        tempJugStats[x] = tempJugStats[x] - waterCanGive
                        
                    tempTuple = (tempJugStats[0], tempJugStats[1], tempJugStats[2])
                    
                    #add the new jug state to the queue as a node
                    newNode = Node(currNode, tempTuple)
                    inList = False
                    length = len(closedList)
                    for i in range(length):
                        if (tempTuple == closedList[i]):
                            inList = True
                    if inList == False:
                        queue.put(newNode)
                        currNode.children.append(newNode)
                        
                    tempJugStats.clear()
                        
        #add the current node to the closed list - it's now been expanded on
        closedList.append(tupleCurr)
                    
    #print('Left the Loop')
    
    if (tupleCurr != goal and queue.empty() == True):
        print('Solution Not Found')
        
    #create a path list to the goal, if it exists
    else:
        print('Solution Found')
        pathList = []
    
        while currNode.parent != None:
            pathList.append(currNode.data)
            currNode = currNode.parent
    
        #add initial jug values to the list
        pathList.append(currNode.data)
    
        pathList.reverse()
    
        #print the path taken from the start to the goal
        length = len(pathList)
        for i in range(length):
            print(pathList[i])
            
    print('This BFS search took ', stepCounter, ' steps to find a solution pathway')

```
